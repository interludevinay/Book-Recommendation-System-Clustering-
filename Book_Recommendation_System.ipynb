```python
rating_with_book = rating.merge(book,on='ISBN')
```

```python
num_rating = rating_with_book.groupby('title')['rating'].count()
```

```python
num_rating
```

```python
final_rating = rating_with_book.merge(num_rating,on='title')
```

```python
final_rating
```

```python
final_rating.rename(columns={'rating_y':'no_of_rating','rating_x':'rating'},inplace=True)
```

```python
final_rating
```

```python
final_rating['no_of_rating'].unique()
```

```python
final_rating = final_rating[final_rating['no_of_rating']>50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[240,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[[i]]])
```

```python
def recommend_book(book_name):
  book_id = np.where(book_pivot.index==book_name)[0][0]
  distances,suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1),n_neighbors=6)

  for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[[i]]])
```

```python
recommend_book('Hideaway')
```
